### Let's learn about two keywords:

- continue

- break

```python
start = -2
end = 2
while (start<end):
    if(start==0):
        start+=1
        continue
    
    print(100/start)
    start +=1
```

```python
start = 0
end = 4

while (start<end):
    print(start+1)
    start+=1
    if(start%2==0):
        continue
```
